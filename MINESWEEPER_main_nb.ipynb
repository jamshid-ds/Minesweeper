import **pandas** and **numpy**


```python
import numpy as np
import pandas as pd
```

Create DataFrame


```python
data = {
    "x":[0,0,0],
    "y":[0,1,2]
}

df = pd.DataFrame(data)
df
```

Create a new column of **`zeros`** (safe) and **`ones`** (mine) for this DataFrame df. The probability of a mine occurring at each location was 0.4.

```python
np.random.seed(42)
df['safe'] = np.random.choice([0, 1], size=len(df), p=[0.6, 0.4])
df['mine'] = np.random.choice([0, 1], size=len(df), p=[0.6, 0.4])
df
```

Now a new **column** called `adjacent` is created for this `DataFrame`. This column contains the number of mines found in adjacent squares on the grid.
(For example, for the first line with an entry for the coordinate (0, 0), the number of mines at the coordinates `(0, 1)`, `(1, 0)` and `(1)` is counted, 1).)

```python
df['adjacent'] = df.apply(lambda row: sum((row['x']==df['x']+1)&(row['y']==df['y']) |  (row['x'] == df['x'] - 1) & (row['y'] == df['y']) |
                                          (row['x'] == df['x']) & (row['y'] == df['y'] + 1) | (row['x'] == df['x']) & (row['y'] == df['y'] - 1)), axis=1)
df
```

Changed value in `adjacent` column to `NaN` for **DataFrame** rows containing mine

```python
df['adjacent']=np.where(df['adjacent']>0, df['adjacent'], np.nan)
df
```

Convert the DataFrame to a grid of the number of `adjacent` mines: columns `x`-coordinate, rows `y`-coordinate.

```python
df['adjacent'] = df.pivot(index='y', columns='x', values='adjacent')
df
```
